# Display examples
This notebook shows some examples on how to show data with the `jupyternimpkg/display` package

A markdown file:

```nim
import jupyternimpkg/display
show dkTextFile:
    "mdfile.md"
```

A `png`:

```nim
import jupyternimpkg/display
show dkImageFile: 
    "pngfile.png"
```

A `.gif`

```nim
import jupyternimpkg/display
# a desired image size can be specified too
show dkImageFile, [320,240]: 
    "giffile.gif"
```

Print some latex, also with interpolations:

```nim
latex "x = \\frac{1}{2}"
import math, strutils
let x = if 5>0: fac(5) else: 0
latex """ x = \begin{cases} 5!, & \mbox{if } 5>0 \\ 0, & \mbox{otherwise} \end{cases} = $1 """ % $x
```

Similar `proc`s are available with `html`, `md` , and they can live in the same codecell

```nim
import jupyternimpkg/display

md "# Mixing stuff"

latex "x = \\frac{1}{2}=", 1/2

echo "hi"
```
